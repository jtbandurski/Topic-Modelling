```python
import selenium 
from selenium import webdriver
import time

from webdriver_manager.chrome import ChromeDriverManager 

import requests 
from bs4 import BeautifulSoup

import pickle

def save_object(obj, filename):
    with open(filename, 'wb') as output:
        pickle.dump(obj, output, pickle.HIGHEST_PROTOCOL)

import pandas as pd 
import re
```

```python
chromepath = ChromeDriverManager().install()
print(chromepath)
```

```python
websites_list = []
for i in range(1,38): #from April 1, 2022 there were 370 news
    websites_list.append("https://us.cnn.com/search?q=world+cup+qatar&from="+str((i-1)*10)+"&size=10&page="+str(i)+"&sort=newest&types=article&section=")

dates_xpaths_list = [] 
texts_xpaths_list = []
for i in range(1,11):
    dates_xpaths_list.append("//body[@class='layout layout-no-rail cnn']/div[@class='layout__content-wrapper layout-no-rail__content-wrapper']/section[@class='layout__wrapper layout-no-rail__wrapper']/section[@class='layout__main layout-no-rail__main']/div[@class='static']/section[@class='tabcontent active']/section[@class='search']/div[@id='search']/div[@class='search__right']/div[@class='search__results']/div[@class='search__results-list']/div["+str(i)+"]/a[1]/div[2]/div[2]")
    texts_xpaths_list.append("/html[1]/body[1]/div[1]/section[3]/section[1]/div[1]/section[1]/section[1]/div[1]/div[2]/div[1]/div[2]/div["+str(i)+"]/a[1]/div[2]/div[3]")
    
cookies_button_xpath = '''//*[@id="onetrust-accept-btn-handler"]'''
```

```python
data = []
i = 1
for website in websites_list:
        
        if(i%25==0):
            print('website:',i,'out of',len(websites_list))
        i=i+1
            
        driver = webdriver.Chrome(chromepath)

        chrome_options = webdriver.ChromeOptions()
        chrome_options.add_argument("--disable-popup-blocking")

        driver.maximize_window()
        driver.get(website) 
        
        time.sleep(5)

        content = driver.find_element("xpath",cookies_button_xpath)
        content.click()

        for j in range(0,10):
            
            try:
                date = driver.find_element("xpath", dates_xpaths_list[j]) 
                element = driver.find_element("xpath", texts_xpaths_list[j])

                data.append([date.text,element.text,'cnn'])
                
            except:
                continue
                
        driver.close()
```

```python
data_df = pd.DataFrame(data)
data_df
```
